```python
import csv
import json

# File names
input_csv_file = 'products.csv'
output_json_file = 'data.json'

# Read the CSV and add data to a list of dictionaries
with open(input_csv_file, mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    data = list(csv_reader)

# Write the list to a JSON file
with open(output_json_file, mode='w', encoding='utf-8') as json_file:
    json.dump(data, json_file, indent=4)

print(f"✅ Successfully converted '{input_csv_file}' to '{output_json_file}'")
```

```python
import json

# File to validate
json_file = 'data.json'

# Load JSON data
with open(json_file, mode='r', encoding='utf-8') as file:
    data = json.load(file)

# Check each record for null/empty values
print("🔍 Checking for null or empty fields...\n")
for item in data:
    for key, value in item.items():
        if value is None or value == "":
            print(f"❗ Record ID {item.get('id', 'N/A')} → Field '{key}' is empty or null.")

print("\n✅ Validation complete.")
```

```python
import json

# File to check
json_file = 'data.json'

# Load JSON data
with open(json_file, mode='r', encoding='utf-8') as file:
    data = json.load(file)

# Count null/empty image_url fields
missing_count = 0
for item in data:
    if 'image_url' not in item or item['image_url'] is None or item['image_url'] == "":
        missing_count += 1

print(f"🖼️ Total records with missing 'image_url': {missing_count}")
```

```python
import json

# File names
input_json_file = 'data.json'
cleaned_json_file = 'data_with_nonempty_image.json'

# Load original JSON data
with open(input_json_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Filter: remove products with empty or missing image field
filtered_data = [
    item for item in data
    if item.get('image_url') not in (None, "", "null")
]

# Save cleaned data
with open(cleaned_json_file, 'w', encoding='utf-8') as f:
    json.dump(filtered_data, f, indent=4)

print(f"✅ Cleaned file saved as '{cleaned_json_file}' with {len(filtered_data)} valid products.")
```

```python
import json
from collections import Counter

# File to read
json_file = 'data_with_nonempty_image.json'

# Load JSON
with open(json_file, mode='r', encoding='utf-8') as file:
    data = json.load(file)

# Extract all names
names = [item.get('name', '').strip() for item in data if 'name' in item and item['name'].strip() != ""]

# Count duplicates
name_counts = Counter(names)
duplicates = {name: count for name, count in name_counts.items() if count > 1}

# Output
print(f"🔁 Total duplicate names: {len(duplicates)}\n")
for name, count in duplicates.items():
    print(f"• '{name}' → {count} times")
```

```python
import json

# Input and output files
input_file = 'data_with_nonempty_image.json'
output_file = 'data_with_no_duplicate_names.json'

# Load JSON
with open(input_file, mode='r', encoding='utf-8') as file:
    data = json.load(file)

# Track names we've already seen
seen_names = set()
unique_data = []

# Keep only the first occurrence of each name
for item in data:
    name = item.get('name', '').strip()
    if name and name not in seen_names:
        seen_names.add(name)
        unique_data.append(item)

# Save cleaned data to new file
with open(output_file, mode='w', encoding='utf-8') as out_file:
    json.dump(unique_data, out_file, indent=4)

print(f"✅ Duplicate names removed. First occurrences saved to '{output_file}'.")
print(f"📦 Total unique records: {len(unique_data)}")
```

```python
import json

# Input and output file names
input_file = 'data_with_no_duplicate_names.json'
output_file = 'data_with_no_duplicate_names_with_ids.json'

# Load JSON data
with open(input_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Add an 'id' field to each object
for idx, item in enumerate(data, start=1):
    item['id'] = idx

# Save updated data
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4)

print(f"✅ IDs added and saved to '{output_file}'")
```

Concatenating 2 jsons

```python
import json

# Load both files
with open('data2_with_no_duplicate_names.json', 'r', encoding='utf-8') as f1:
    data1 = json.load(f1)

with open('data_with_no_duplicate_names.json', 'r', encoding='utf-8') as f2:
    data2 = json.load(f2)

# Combine the lists
combined_data = data1 + data2

# Save to a new file
with open('combined_data.json', 'w', encoding='utf-8') as fout:
    json.dump(combined_data, fout, indent=4)

print("✅ Combined JSON saved to 'combined.json'")
```
